```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df_heart = pd.read_csv('./heart_disease.data',
                      header=None)
df_adultincome = pd.read_csv('./adult_income.data',
                      header=None)
df_redwine = pd.read_csv('./winequality-red.csv', delimiter=';')
df_whitewine = pd.read_csv('./winequality-white.csv', delimiter=';')
```

```python
plt.bar(df_heart[13].value_counts().keys(), df_heart[13].value_counts())

plt.ylabel('Occurrences')
plt.xlabel('Categories')
plt.title('Presence of a heart disease')
plt.savefig('heart_disease.png', dpi=300)
plt.show()
```

```python
plt.bar(['heart', 'income', 'wine quality'], [df_heart.shape[0], df_adultincome.shape[0], df_whitewine.shape[0] + df_redwine.shape[0]])

plt.ylabel('instances')
plt.xlabel('Data sets')
plt.savefig('datasets.png', dpi=300)
plt.show()
```

```python
plt.bar(df_adultincome[14].value_counts().keys(), df_adultincome[14].value_counts())

plt.ylabel('Occurrences')
plt.xlabel('Categories')
plt.title('50k yearly earning')
plt.savefig('adult_income.png', dpi=300)
plt.show()
```

```python
plt.bar(df_redwine['quality'].value_counts().keys(), df_redwine['quality'].value_counts(), alpha=0.5, width=0.25, color='red', label='red wine')
plt.bar(df_whitewine['quality'].value_counts().keys() + 0.25, df_whitewine['quality'].value_counts(), alpha=0.5, width=0.25,  color='yellow', label='white wine')
plt.ylabel('Occurrences')
plt.xlabel('Quality')
plt.title('wine quality')
plt.legend()
plt.savefig('wine_quality.png', dpi=300)
plt.show()
```
